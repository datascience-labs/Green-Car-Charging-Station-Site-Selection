```python
def get_location(addr):
    print(addr)
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + addr
    headers = {"Authorization": "KakaoAK 57b91d36b9c73f63c09809b49093eb07"}
    api_json = json.loads(str(requests.get(url,headers=headers).text))
    if len(api_json['documents']) == 0:
        x_y = [0, 0]
        return x_y
    address = api_json['documents'][0]['address']
    if address is not None:
        x_y = [str(address['y']),str(address['x'])]
        print(x_y)
        return x_y
     
    else:
        x_y = [0, 0]
        return x_y
```

```python
lat = []
long = []
for i in addr:
    lat.append(get_location(i)[0])
    long.append(get_location(i)[1])
```
